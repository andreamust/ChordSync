```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "genre": [
        "Rock",
        "Pop",
        "Folk",
        "Country",
        "Religious Music",
        "R&b, Funk & Soul",
        "Soundtrack",
        "World Music",
        "Electronic",
        "Metal",
        "Hip Hop",
        "Jazz",
        "Reggae & Ska",
        "Classical",
        "Comedy",
        "Blues",
        "Disco",
        "Experimental",
        "New Age",
        "Darkwave",
    ],
    "count": [
        454128,
        148142,
        139728,
        89670,
        46372,
        35899,
        31876,
        25776,
        20851,
        18328,
        17895,
        7187,
        6734,
        5542,
        4279,
        3447,
        1561,
        540,
        529,
        399,
    ],
}
```

```python
# plot data
df = pd.DataFrame(data)
df.set_index("genre", inplace=True)
df.plot(kind="bar", figsize=(10, 6))

plt.title("Number of songs by genre")
plt.ylabel("Number of songs")
plt.xlabel("Genre")

plt.show()
```

```python
# plot it logarithmically
df.plot(kind="bar", figsize=(10, 6), logy=True)

plt.title("Number of songs by genre (logarithmic)")
plt.ylabel("Number of songs")
plt.xlabel("Genre")

plt.show()
```
