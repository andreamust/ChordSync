```python
# plot it logarithmically
df.plot(kind="bar", figsize=(10, 6), logy=True)

plt.title("Number of songs by genre (logarithmic)")
plt.ylabel("Number of songs")
plt.xlabel("Genre")

plt.show()
```

```python
# create a log bar plot comparing data from both datasets
df_1 = pd.DataFrame(data_ultimate)
df_1.set_index("genre", inplace=True)

df_2 = pd.DataFrame(data_choco)
df_2.set_index("genre", inplace=True)

fig, ax = plt.subplots(figsize=(18, 6))
df_1.plot(kind="bar", ax=ax, color="#E8AD05", position=0, width=0.4, logy=True)
df_2.plot(kind="bar", ax=ax, color="black", position=1, width=0.4, logy=True)

# make labels font bigger
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.legend(["Ultimate Guitar", "ChoCo"], fontsize=15)
plt.title("Number of songs by genre", fontsize=20)
plt.ylabel("Number of songs", fontsize=20)
plt.xlabel("Genre", fontsize=20)

# set background color
plt.gca().set_facecolor("#F4F4F4")
fig.patch.set_facecolor("#F4F4F4")

plt.show()
```
